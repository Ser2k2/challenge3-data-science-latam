##Encoding

###get_dummies

```python
list(df_2.columns)
```

```python
X = df_2.drop(columns=['churn'], axis= 1)
y = df_2['churn']
```

```python
categoricas = ['gender', 'senior_citizen', 'partner', 'dependents',
               'phone_service', 'multiple_lines', 'internet_service',
               'online_security', 'online_backup', 'device_protection',
               'tech_support', 'streaming_tv', 'streaming_movies',
               'contract', 'paperless_billing','payment_method']
```

####Transformando X

```python
X_dummies = pd.get_dummies(data= X, columns= categoricas, dtype= int)
```

```python
X_dummies.head(5)
```

####Transformando y

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
y = label_encoder.fit_transform(y)
y
```

###OneHotEcnoder

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot_enc = make_column_transformer((OneHotEncoder(handle_unknown='ignore'), categoricas), remainder='passthrough')
```

```python
X_ohe = one_hot_enc.fit_transform(X)
X_ohe = pd.DataFrame(X_ohe, columns=one_hot_enc.get_feature_names_out())
X_ohe.head(5)
```

####SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
over_sampling = SMOTE(random_state= 42)
```

```python
X_resample_smote, y_resample_smote = over_sampling.fit_resample(X_ohe, y)
```

```python
print(sorted(Counter(y_resample_smote).items()))
```

###Under-sampling

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus = RandomUnderSampler(random_state=42)
X_resampled_under, y_resampled_under = rus.fit_resample(X_ohe, y)
```

```python
print(sorted(Counter(y_resampled_under).items()))
```

#ü§ñ Modelado Predictivo

##Separaci√≥n de Datos

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_ohe, y, test_size= 0.3,random_state= 42)
```

##Creaci√≥n de Modelos

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
arbol = DecisionTreeClassifier(max_depth= 3)
random_forest = RandomForestClassifier(max_depth= 2)
```

##Evaluaci√≥n de los Modelos

###Exactitud (Acur√°cia), Precisi√≥n, Recall, F1-score y Matriz de confusi√≥n

```python
arbol.fit(X_train, y_train)
print(f'Exactitud  de entrenamiento (Decision Tree): {arbol.score(X_train, y_train)}')
print(f'Exactitud  de prueba (Decision Tree): {arbol.score(X_test, y_test)}')
```

```python
random_forest.fit(X_train, y_train)
print(f'Exactitud de entrenamiento (Random Forest): {random_forest.score(X_train, y_train)}')
print(f'Exactitud de prueba (Random Forest): {random_forest.score(X_test, y_test)}')
```

```python
from sklearn.metrics import accuracy_score,recall_score, precision_score, f1_score
```

```python
prevision_arbol = arbol.predict(X_test)
```

```python
print(f'Precisi√≥n (Decision Tree): {accuracy_score(y_test, prevision_arbol)}')
print(f'Recall (Decision Tree): {recall_score(y_test, prevision_arbol)}')
print(f'Precisi√≥n(Decision Tree): {precision_score(y_test, prevision_arbol)}')
print(f'F1-Score (Decision Tree): {f1_score(y_test, prevision_arbol)}')
```

```python
prevision_rf = random_forest.predict(X_test)
```

```python
print(f'Precisi√≥n (Random Forest): {accuracy_score(y_test, prevision_rf)}')
print(f'Recall (Random Forest): {recall_score(y_test, prevision_rf)}')
print(f'Precisi√≥n (Random Forest): {precision_score(y_test, prevision_rf)}')
print(f'F1-Score (Random Forest): {f1_score(y_test, prevision_rf)}')
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
prediccion_arbol = arbol.predict(X_test)
ConfusionMatrixDisplay.from_predictions(y_test, prediccion_arbol)
```

```python
prediccion_rf = random_forest.predict(X_test)
ConfusionMatrixDisplay.from_predictions(y_test, prediccion_rf)
```
